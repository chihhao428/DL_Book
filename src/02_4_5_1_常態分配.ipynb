## 常態分配的積分

```python
# 載入套件
import scipy.integrate as integrate
import numpy as np
import math


# 平均數(mean)、標準差(std)
mean = 0
std = 1
# 常態分配的機率密度函數(Probability Density Function, pdf)
f = lambda x: (1/((2*np.pi*std**2) ** .5)) * np.exp(-0.5*((x-mean)/std)**2) 

# 積分，從負無窮大至無窮大
i, e = integrate.quad(f, -np.inf, np.inf)

print('累積機率:', round(i, 2))
print('誤差:', str(e))
```

## 計算常態分配的正負1、2、3倍標準差區間之機率

```python
# 1倍標準差區間之機率
i, e = integrate.quad(f, -1, 1)

print('累積機率:', round(i, 3))
print('誤差:', str(e))
```

```python
# 2倍標準差區間之機率
i, e = integrate.quad(f, -2, 2)

print('累積機率:', round(i, 3))
print('誤差:', str(e))
```

```python
# 3倍標準差區間之機率
i, e = integrate.quad(f, -3, 3)

print('累積機率:', round(i, 3))
print('誤差:', str(e))
```

```python
# 95%信賴區間之機率
i, e = integrate.quad(f, -1.96, 1.96)

print('累積機率:', round(i, 3))
print('誤差:', str(e))
```

## 使用隨機亂數及直方圖，繪製標準常態分配圖

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 使用 randn 產生標準常態分配的亂數 10000 個樣本點
x = np.random.randn(10000)

# 直方圖，參數 hist=False：不畫階梯直方圖，只畫平滑曲線
sns.distplot(x, hist=False)

# 設定圖形屬性
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# 對正負1、2、3倍標準差畫虛線
plt.axvline(-3, c='r', linestyle=':')
plt.axvline(3, c='r', linestyle=':')
plt.axvline(-2, c='g', linestyle=':')
plt.axvline(2, c='g', linestyle=':')
plt.axvline(-1, c='b', linestyle=':')
plt.axvline(1, c='b', linestyle=':')

# 1 倍標準差機率
plt.annotate(text='', xy=(-1,0.25), xytext=(1,0.25), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='68.3%', xy=(0,0.26), xytext=(0,0.26))
# 2 倍標準差機率
plt.annotate(text='', xy=(-2,0.05), xytext=(2,0.05), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='95.4%', xy=(0,0.06), xytext=(0,0.06))
# 3 倍標準差機率
plt.annotate(text='', xy=(-3,0.01), xytext=(3,0.01), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='99.7%', xy=(0,0.02), xytext=(0,0.02))

plt.show()
```

## 範例1. 使用SciPy繪製機率密度函數(pdf)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# 觀察值範圍
z1, z2 = -4, 4

# 樣本點
x = np.arange(z1, z2, 0.001) 
y = norm.pdf(x,0,1)

# 繪圖
plt.plot(x,y,'black')

# 填色
plt.fill_between(x,y,0, alpha=0.3, color='b')
plt.show()
```

## 範例3. 使用SciPy繪製累積分配函數(cdf)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# 觀察值範圍
z1, z2 = -4, 4

# 樣本點
x = np.arange(z1, z2, 0.001) 

# norm.cdf：累積分配函數
y = norm.cdf(x,0,1)

# 繪圖
plt.plot(x,y,'black')

plt.show()
```

## 範例4. 使用SciPy繪製Student’s t 分配的機率密度函數(pdf)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t, norm

# 觀察值範圍
z1, z2 = -4, 4

# 樣本點
x = np.arange(z1, z2, 0.1) 
dof = 10 #len(x) - 1
y = t.pdf(x, dof)

# 繪圖
plt.plot(x,y,'red', label='t分配', linewidth=5)

# 繪製常態分配
y2 = norm.pdf(x)
plt.plot(x,y2, 'green', label='常態分配')

plt.legend()
plt.show()
```
