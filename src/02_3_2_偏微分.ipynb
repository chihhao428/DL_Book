# 偏微分(Partial Differentiation)

## 範例1. 簡單線性迴歸
$\begin{equation}y = 2x + 1\end{equation}$

```python
# 載入套件
import matplotlib.pyplot as plt
import numpy as np
import math

## 建立等距方格(meshgrid)的交叉點
el = np.arange(-5,6)
nx, ny = np.meshgrid(el, el, sparse=False, indexing='ij')

## 建構等高線，計算 z 值
x_coord = []
y_coord = []
z = []
for i in range(11):  
    for j in range(11):
        x_coord.append(float(-nx[i,j]))
        y_coord.append(float(-ny[i,j]))       
        z.append(nx[i,j]**2 + ny[i,j]**2)

## 計算 x、y 梯度(gradients)
x_grad = [-2 * x for x in x_coord]
y_grad = [-2 * y for y in y_coord] 

## 沿著梯度，畫箭頭(arrows)
plt.xlim(-5.5,5.5)
plt.ylim(-5.5,5.5)
for x, y, xg, yg in zip(list(x_coord), list(y_coord), list(x_grad), list(y_grad)):
    if x != 0.0 or y != 0.0: ## Avoid the zero divide when scaling the arrow
        l = math.sqrt(xg**2 + yg**2)/2.0
        plt.quiver(x, y, xg, yg, width = l, units = 'dots')

## 畫等高線(Contour)
z = np.array(z).reshape(11,11)    
plt.contour(el, el, z)    
plt.show()
```

## 函數如下：
### $f(x,y) = x^2 + y^2$

## 偏微分
### $\frac{\partial f(x,y)}{\partial x} = \frac{\partial (x^2 + y^2)}{\partial x} = 2x + 0 = 2x$

### $\frac{\partial f(x,y)}{\partial y} = \frac{\partial (x^2 + y^2)}{\partial x} = 2y + 0 = 2y$



```python
# f(x) = x^2 + y^2

from sympy import *
x, y = symbols('x y')

# 對 x 偏微分
y1 = x**2 + y**2
yprime1 = y1.diff(x)
print(yprime1)

# 對 y 偏微分
yprime2 = y1.diff(y)
print(yprime2)
```

## 利用梯度下降法(Gradient Descent)找最小值

### 1. 可以任意設定一起始點(x_start)
### 2. 計算該點的梯度 fd(x)
### 3. 沿著梯度更新x，逐步逼近最佳解，幅度大小以學習率控制
####     新的x = x — 學習率(learning rate) * 梯度(gradient)
### 4. 重複步驟2、3，判斷梯度是否接近於0，若已很逼近，即找到最佳解


```python
import numpy as np
import matplotlib.pyplot as plt

# 函數 f(x)=x^2
def f(x): return x ** 2 

# 一階導數:dy/dx=2*x
def fd(x): return 2 * x

def GD(x_start, df, epochs, lr):    
    xs = np.zeros(epochs+1)    
    w = x_start    
    xs[0] = w    
    for i in range(epochs):         
        dx = df(w)        
        # 權重的更新 
        # W_NEW = W — 學習率(learning rate) x 梯度(gradient)
        w += - lr * dx         
        xs[i+1] = w    
    return xs


# 超參數(Hyperparameters)
x_start = 5    # 起始權重
epochs = 25    # 執行週期數
lr = 0.1       # 學習率

# 梯度下降法, 函數 fd 直接當參數傳遞 
w = GD(x_start, fd, epochs, lr=lr) 
# 顯示每一執行週期得到的權重
print (np.around(w, 2))

# 畫圖
color = 'r'    
from numpy import arange
t = arange(-6.0, 6.0, 0.01)
plt.plot(t, f(t), c='b')
plt.plot(w, f(w), c=color, label='lr={}'.format(lr))    
plt.scatter(w, f(w), c=color, ) 

# 設定中文字型
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"c:\windows\fonts\msjhbd.ttc", size=20)   
plt.title('梯度下降法', fontproperties=font)
plt.xlabel('w', fontsize=20)
plt.ylabel('Loss', fontsize=20)

# 矯正負號
plt.rcParams['axes.unicode_minus'] = False

plt.show()
```
